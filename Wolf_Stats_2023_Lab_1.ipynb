```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

```python
y = -3*x-12
```

```python
plt.plot (x,y, label = "-3x-12")
plt.title("Graph of y=-3x-12");
plt.xlabel("x");
plt.ylabel("y");
plt.legend();
```

```python
y = np.sin(x)
```

```python
plt.plot (x,y, label = "sin(x)")
plt.title("Graph of y=sin(x)");
plt.xlabel("x");
plt.ylabel("y");
plt.legend();
```

```python
y = np.exp(x)
```

```python
plt.plot (x,y, label = "exp(x)")
plt.title("Graph of y=exp(x)");
plt.xlabel("x");
plt.ylabel("y");
plt.legend();
```

```python
y = np.exp(-x**2)
```

```python
plt.plot (x,y, label = "exp(x^2)")
plt.title("Graph of y=x^2)");
plt.xlabel("x");
plt.ylabel("y");
plt.legend();
```

```python
labels = '4 dawgs','6 dawgs','2 dawgs','5 dawgs', 'other'
sizes = [45, 15, 12, 8, 20]
explode = (0.1, 0, 0, 0, 0)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.title("Possible amt. of Hot Dogs Eaten in One Sitting")
plt.show()
```

This graph was produced using data collected from my friends over the course of two days. It is important to note that the numbers reflected here represent the **maximim** amount of hot dogs eaten in one sitting; **not** how many they typically eat.
