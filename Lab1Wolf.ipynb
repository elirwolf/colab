```python
x = np.linspace(0,10,1000) #start value, stop value, how many steps
```

```python
x
```

```python
y = x**2
```

```python
plt.plot(x,y, marker = 'd',
         linestyle = 'none',
         markerfacecolor= '#c4b6db')
```

```python
plt.plot(x,y, marker = 'd', linestyle = 'none', markerfacecolor= '#c4b6db')
```

```python
plt.plot(x,y, marker = 'd', linestyle = 'none', markerfacecolor= '#c4b6db')
plt.title("My Fisrt Graph");
```

```python
plt.plot(x,y, marker = 'd', linestyle = '--', markerfacecolor= '#b8e69e')
plt.title("My Fisrt Graph");
plt.xlabel("x axis");
plt.ylabel("y axis");
# put the ; if you dont want all the extra stuff
```

```python
plt.plot(x,y, marker = '*', linestyle = '--', markerfacecolor= '#b8e69e', label = "Graph of x Squared")
plt.title("My Fisrt Graph");
plt.xlabel("x axis");
plt.ylabel("y axis");
plt.legend();
# put the ; if you dont want all the extra stuff
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x, label="linear")
plt.plot(x,x**2, label="squared")
plt.plot(x,x**3, label="cubed")
plt.legend();
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x, 'm--', label="linear",)
plt.plot(x,x**2, 'c*', label="squared")
plt.plot(x,x**3, 'ko', label="cubed")
plt.legend();
```
